#### Investigating the numerical stability of Softmax 

```python
import numpy as np 
import torch 
```

```python
def naive_softmax(logits):
    exp_logits = np.exp(logits)
    return exp_logits/np.sum(exp_logits)
def shifted_softmax(logits):
    ''' 
    subtract each element from the largest value to normalize 
    '''
    C= torch.max(logits) 
    logits = logits -C 
    return torch.softmax(logits, dim=-1)
```

```python
logits =np.array([1000.0, 1001.0, 999.0])
unstable = naive_softmax(logits)
print(unstable)
```

```python
logits =torch.tensor([1000.0, 1001.0, 999.0])
stable = shifted_softmax(logits)
stable
```
