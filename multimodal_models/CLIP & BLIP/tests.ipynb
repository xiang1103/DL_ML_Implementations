```python
import torch
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt

def visualize_transform(image_path):
    # 1. Load the original image
    orig_img = Image.open(image_path)

    width, height = orig_img.size 
    # 2. Define the transformation pipeline
    # We resize the smaller edge to 256, then crop to 224
    transform_pipeline = transforms.Compose([
        transforms.Resize((224)),
        #transforms.CenterCrop(224),
        transforms.ToTensor()
    ])

    # 3. Apply transformation
    transformed_tensor = transform_pipeline(orig_img)

    # 4. Prepare for visualization
    # Convert tensor back to HWC format for Matplotlib: (C, H, W) -> (H, W, C)
    transformed_img = transformed_tensor.permute(1, 2, 0).numpy()

    print(transformed_img.shape)

    # 5. Plotting
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))

    # Show Original
    ax[0].imshow(orig_img)
    ax[0].set_title(f"Original\n{orig_img.size[0]}x{orig_img.size[1]}")
    ax[0].axis('off')

    # Show Transformed
    ax[1].imshow(transformed_img)
    ax[1].set_title(f"Transformed Resized")
    ax[1].axis('off')

    plt.tight_layout()
    plt.show()

# Run it
visualize_transform("../training_materials/images/bicycle_woman.jpg")
```
